```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
```

```python
vectorizer = TfidfVectorizer(stop_words = stop_words, min_df = 0.1)
```

```python
tfidf = vectorizer.fit_transform(df['Lyrics'])
```

```python
from sklearn.decomposition import NMF
```

```python
nmf = NMF(n_components = 6)
```

```python
topic_values = nmf.fit_transform(tfidf)
```

```python
for topic_num, topic in enumerate(nmf.components_):
    message = "Topic #{} ".format(topic_num + 1)
    message += " ".join([vectorizer.get_feature_names()[i] for i in topic.argsort()[:-20 :-1]])
    print(message)
```

```python
topic_labels = ['Introspection','Love','Filler','Romance','Joy','Life']
```

```python
topic_values =
```

```python
df_topics = pd.DataFrame(topic_values, columns = topic_labels)
df_topics
```

```python
df = df.join(df_topics)
```

```python
df.head()
```

```python
df.loc[df['Introspection'] >= 0.1,'Introspection'] = 1
df.loc[df['Love'] >= 0.1,'Love'] = 1
df.loc[df['Filler'] >= 0.1,'Filler'] = 1
df.loc[df['Romance'] >= 0.1,'Romance'] = 1
df.loc[df['Joy'] >= 0.1,'Joy'] = 1
df.loc[df['Life'] >= 0.1,'Life'] = 1
```

```python
df.loc[df['Introspection'] < 0.1,'Introspection'] = 0
df.loc[df['Love'] < 0.1,'Love'] = 0
df.loc[df['Filler'] < 0.1,'Filler'] = 0
df.loc[df['Romance'] < 0.1,'Romance'] = 0
df.loc[df['Joy'] < 0.1,'Joy'] = 0
df.loc[df['Life'] < 0.1,'Life'] = 0
```

```python
album_topics = df.groupby('Album').sum().reset_index()
```
